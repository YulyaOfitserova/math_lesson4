```python
#решить графически sin(a*x) = 0, 0,01<a<0,02, 100<x<500
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

for k in range(1, 4):
    a = np.linspace(0.01, 0.02, 20)
    plt.plot(a, k * np.pi/a)      
    
plt.xlabel('a')
plt.ylabel('x')
plt.ylim(100, 500)
plt.grid(True)
plt.show()
```
